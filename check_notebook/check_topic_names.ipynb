```python
import pandas as pd
import numpy as np

topic_names_df = pd.read_csv("outputs/topic_names/topic_names.csv")
df_raw = pd.read_excel("data/bulletin_data.xlsx")
df_sub = pd.read_csv("outputs/subtopics/data_with_subtopics.csv")
```

```python
topic_names_df.head()
```

```python
df_sub['topic_name'] = df_sub['topic'].map(topic_names_df.set_index('topic')['topic_name'])

# 3. 매핑 결과 확인
print(f"전체 데이터 수: {len(df_sub)}")
print(f"topic_name이 매핑된 데이터 수: {df_sub['topic_name'].notna().sum()}")
print(f"매핑되지 않은 데이터 수: {df_sub['topic_name'].isna().sum()}")

# 4. 매핑되지 않은 topic 값들 확인 (있다면)
unmapped_topics = df_sub[df_sub['topic_name'].isna()]['topic'].unique()
if len(unmapped_topics) > 0:
    print(f"매핑되지 않은 topic 값들: {unmapped_topics}")

# 6. topic_name별 데이터 수 확인
print("\n=== Topic별 데이터 수 ===")
topic_counts = df_sub['topic_name'].value_counts()
print(topic_counts.head(10))

# 5. 결과 확인 (상위 5개 행)
print("\n=== 매핑 결과 확인 ===")
print("topic 수:", len(df_sub['topic'].unique()))
df_sub.head(2)
```

```python
df_sub.to_csv("outputs/topic_names/data_with_topic_names.csv", index=False)
```
