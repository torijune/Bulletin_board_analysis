```python
import pandas as pd
import re

df = pd.read_csv("/Users/jang-wonjun/Desktop/Dev/Bulletin_board_analysis/outputs/analysis/final_analysis/topic_analysis_results.csv")
major_topic_names_df = pd.read_csv("/Users/jang-wonjun/Desktop/Dev/Bulletin_board_analysis/outputs/major_topic_names/major_topic_names.csv")
topic_names_df = pd.read_csv("/Users/jang-wonjun/Desktop/Dev/Bulletin_board_analysis/outputs/topic_names/topic_names.csv")
```

```python
import re
import pandas as pd

def extract_count_from_subtopic(subtopic):
    """소주제에서 건수를 추출합니다."""
    try:
        match = re.search(r'\((\d+)건\)', subtopic)
        if match:
            return int(match.group(1))
        else:
            return 0
    except:
        return 0

def extract_subtopic_name(subtopic):
    """소주제에서 건수 부분을 제거한 이름을 추출합니다."""
    try:
        name = re.sub(r'\s*\(\d+건\)\s*$', '', subtopic)
        return name.strip()
    except:
        return subtopic

def process_dataframe(df):
    """데이터프레임을 후처리합니다."""
    if df.empty:
        return df
    
    # 건수 추출
    df['소주제_건수'] = df['소주제'].apply(extract_count_from_subtopic)
    
    # 소주제명 정리
    df['소주제'] = df['소주제'].apply(extract_subtopic_name)
    
    # 대주제별 건수 합계 추가
    df['대주제_건수'] = df.groupby('대주제')['소주제_건수'].transform('sum')
    
    # 컬럼 순서 재정렬
    columns_order = ['대주제', '대주제_건수', '소주제', '소주제_건수', '핵심', '상담예시']
    df = df[columns_order]
    
    return df
```

```python
df = process_dataframe(df)
```

```python
total_sub = df['소주제_건수'].sum()

# 2) 대주제별 건수 합
total_major = df.groupby('대주제')['소주제_건수'].sum()

# 3) 전체 대주제 건수 (unique하게 한 번만 합산할 경우)
total_major_unique = df.drop_duplicates(subset=['대주제'])['대주제_건수'].sum()

print("전체 소주제 건수 합:", total_sub)
print("대주제별 소주제 건수 합:\n", total_major)
print("전체 대주제 건수 합(중복 제거):", total_major_unique)
```

```python
import re
import pandas as pd

def clean_subtopic_name(x: str) -> str:
    """'소주제'에서 '(123건)' 꼬리표 제거 + 공백 정리"""
    if pd.isna(x):
        return x
    return re.sub(r"\s*\(\d+건\)\s*$", "", str(x)).strip()

def attach_topic_ids(df: pd.DataFrame,
                     major_topic_names_df: pd.DataFrame,
                     topic_names_df: pd.DataFrame,
                     major_col: str = "대주제",
                     minor_col: str = "소주제",
                     out_major_col: str = "major_topic_id",
                     out_minor_col: str = "minor_topic_id") -> pd.DataFrame:
    """
    1) 소주제 이름 -> minor_topic_id (topic_names_df) 매핑
    2) 대주제 이름 -> major_topic_id (major_topic_names_df) 매핑
    3) 2에서 Nan이면, 1의 minor_topic_id에서 '앞자리'를 split하여 major_topic_id 보완
       예: minor_topic_id == '5_3' -> major_topic_id = '5'
    """
    res = df.copy()

    # 0) 정리
    res[minor_col + "_clean"] = res[minor_col].apply(clean_subtopic_name)
    res[major_col + "_clean"] = res[major_col].astype(str).str.strip()

    # 1) dict 준비 (양쪽 모두 strip)
    major_map = dict(zip(
        major_topic_names_df["major_topic_name"].astype(str).str.strip(),
        major_topic_names_df["major_topic"].astype(str)  # 문자열로 보관(혼합형 방지)
    ))
    minor_map = dict(zip(
        topic_names_df["topic_name"].astype(str).str.strip(),
        topic_names_df["topic"].astype(str)              # '5_3' 같은 형태
    ))

    # 2) 소주제 이름 -> minor_topic_id
    res[out_minor_col] = res[minor_col + "_clean"].map(minor_map)

    # 3) 대주제 이름 -> major_topic_id (이름 매핑 1차)
    res[out_major_col] = res[major_col + "_clean"].map(major_map)

    # 4) 보완: major_topic_id가 비어 있으면, minor_topic_id에서 앞자리 추출
    need_fill = res[out_major_col].isna() & res[out_minor_col].notna()
    res.loc[need_fill, out_major_col] = (
        res.loc[need_fill, out_minor_col].str.split("_").str[0]
    )

    # 5) 형 변환(선택): 모두 숫자로 바꾸고 싶으면 주석 해제
    # for col in (out_major_col, out_minor_col):
    #     res[col] = pd.to_numeric(res[col], errors="ignore")

    # 6) 매핑 안 된 케이스 확인용 컬럼
    res["__major_unmatched__"] = res[out_major_col].isna()
    res["__minor_unmatched__"] = res[out_minor_col].isna()

    return res

# -------------------------
# 사용 예시
# -------------------------
df = attach_topic_ids(df, major_topic_names_df, topic_names_df,
                      major_col="대주제", minor_col="소주제",
                      out_major_col="major_topic_id", out_minor_col="minor_topic_id")
df[["대주제","소주제","major_topic_id","minor_topic_id"]].head()
```
