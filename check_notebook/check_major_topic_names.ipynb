```python
columns = ['상담일자', '상담유형', '상담요약', '상담인 유형', '상담내용', '연도', 'major_topic_name', 'topic_name', 'major_topic', 'subtopic', 'topic']

df_final = data_with_topic_names[columns]
df_final.head(2)
```

```python
# 상담일자 문자열 정리 (앞부분만)
df_final["상담일자_clean"] = df_final["상담일자"].astype(str).str.split("-").str[0]

# 연, 월 추출
df_final["상담연도"] = df_final["상담일자_clean"].str.extract(r"^(\d{4})").astype("Int64")
df_final["상담월"] = df_final["상담일자_clean"].str.extract(r"^\d{4}\.(\d{1,2})").astype("Int64")

# 분기 계산
df_final["상담분기"] = ((df_final["상담월"] - 1) // 3 + 1).astype("Int64")

# 계절 매핑 함수
def get_season(month):
    if pd.isna(month):
        return pd.NA
    if month in [3, 4, 5]:
        return "봄"
    elif month in [6, 7, 8]:
        return "여름"
    elif month in [9, 10, 11]:
        return "가을"
    else:  # 12, 1, 2
        return "겨울"

df_final["상담계절"] = df_final["상담월"].apply(get_season)

# 확인
df_final[["상담일자", "상담연도", "상담월", "상담분기", "상담계절"]].head(10)
```

```python
# 상담내용이 결측치(NaN)인 행 제거
df_final = df_final.dropna(subset=["상담내용"]).reset_index(drop=True)

print("상담 갯수 (결측치 제거 후):", len(df_final))
print("결측치 갯수:", df_final.isna().sum())
```

```python
final_columns = ['상담일자', '상담유형', '상담요약', '상담인 유형', '상담내용', '연도', 'major_topic_name', 'topic_name', 'major_topic', 'subtopic', 'topic', '상담연도', '상담월', '상담일', '상담분기', '상담계절']
df_final = df_final[final_columns]
print("상담 갯수: ", len(df_final))
print("최종 데이터셋 컬럼:",df_final.columns)
df_final.head(1)
```

```python
df_final.to_csv("outputs/final_data/final_data.csv", index=False)
```
